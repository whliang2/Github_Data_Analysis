```python
df_without_pos_neg_label = df.loc[:, df.columns != 'Good_or_Bad']
df_without_pos_neg_label.head()
describe_table_new = df_without_pos_neg_label.describe()
describe_table_new.loc['cv'] = describe_table_new.loc['std'] / describe_table_new.loc['mean']
df_normal_new = (describe_table_new - describe_table_new.loc['std'])/describe_table_new.loc['mean']
df_normal_new
```

```python
df_normal = (df_without_pos_neg_label - describe_table_new.loc['std'])/describe_table_new.loc['mean']
# df_normal  = df
df_normal.head()

# print(df.iloc[:,4])
df_final = pd.concat([df_normal, df.iloc[:,4]], axis=1)
df_final.head()
```

```python
repo_name_df = repo_name.to_frame()
normalize_nparray = df_final.to_numpy()
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(normalize_nparray)
kmeans.labels_
print("OK")
```

```python

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(kmeans.labels_)
# print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
# a = onehot_ecoded.to_dataframe()
# invert first example
# inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
# print(inverted)
```

```python
df_final['Label'] = onehot_encoded.tolist()
# print(df_final)
df_new = df_final.iloc[:,:6]
# print(df_new)
df_new_np = df_new.to_numpy()
```

```python
x_train, x_test,y_train,y_test = train_test_split(df_new_np,onehot_encoded, test_size =0.8)
x_train.shape
# sparse_categorical_crossentropy
```

```python

model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(4, activation='softmax')
])

# tf.keras.layers.Dropout(0.2)

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=30, verbose = 1)
result = model.evaluate(x_test, y_test, verbose =0)
print("Lost : " + str(result[0]))
print("Accuracy : " + str(result[1]))
```

```python
# Train the KNN Model

classifier = KNeighborsClassifier(n_neighbors=10)
classifier.fit(x_train, y_train)



# Make Prediction with the KNN Model
y_pred = classifier.predict(x_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
